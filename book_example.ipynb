```python
#we have to make a list of names of all the books in the website with 2 star rating 
import requests
import bs4

```

```python
base_url='https://books.toscrape.com/catalogue/page-{}.html'

two_star_books=[]
for n in range(1,51):#because there are 50 pages 
    scrap_url=base_url.format(n)
    res=requests.get(scrap_url)
    soup=bs4.BeautifulSoup(res.text,'lxml')
    books=soup.select('.product_pod')
    for book in books:
        if len(book.select('.star-rating.Two'))!=0:
            book_title=book.select('a')[1]['title']
            two_star_books.append(book_title)
```
